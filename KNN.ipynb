```python
import zipfile
import pandas as pd
import matplotlib.pyplot as plt


ruta_archivo_zip = '/content/drive/MyDrive/abalone.zip'


archivo_datos = 'abalone.data'


with zipfile.ZipFile(ruta_archivo_zip, 'r') as zip_ref:
    with zip_ref.open(archivo_datos) as file:
        # Leemos los datos del archivo y decodificarlos
        datos = file.read().decode('utf-8')

# Dividimos los datos por saltos de línea y comas
filas = [fila.split(',') for fila in datos.strip().split('\n')]

# Creamos un DataFrame de pandas
df = pd.DataFrame(filas, columns=['Sexo', 'Longitud', 'Diámetro', 'Altura', 'Peso_entero', 'Peso_manzana', 'Peso_vísceras', 'Peso_concha', 'Anillos'])

# Mostramos el DataFrame
print(df)

# Graficamos la distribución de la longitud de los abalones
df.info()
```

```python
# Verificamos si hay valores faltantes
print(df.isnull().sum())

# Si hay valores faltantes, eliminar filas con valores faltantes
df = df.dropna()
```

```python
# Convertimos columnas numéricas a tipo numérico
df['Longitud'] = pd.to_numeric(df['Longitud'])
df['Diámetro'] = pd.to_numeric(df['Diámetro'])
df['Altura'] = pd.to_numeric(df['Altura'])
df['Peso_entero'] = pd.to_numeric(df['Peso_entero'])
df['Peso_manzana'] = pd.to_numeric(df['Peso_manzana'])
df['Peso_vísceras'] = pd.to_numeric(df['Peso_vísceras'])
df['Peso_concha'] = pd.to_numeric(df['Peso_concha'])
df['Anillos'] = pd.to_numeric(df['Anillos'])
df
```

```python
# Codificamos la columna "Sexo" usando one-hot encoding
df = pd.get_dummies(df, columns=['Sexo'])
df
```

```python
# Exploramos datos atípicos utilizando describe()
print(df.describe())

# Filtramos datos atípicos basados en umbrales específicos
# Por ejemplo, para eliminar filas con valores de Longitud extremadamente altos
df = df[df['Longitud'] < 1.0]  # Ajusta el umbral según sea necesario

# También podrías aplicar técnicas más avanzadas para identificar y manejar outliers
```

```python
from sklearn.preprocessing import StandardScaler

# Inicializaremos el escalador
scaler = StandardScaler()

# Normalizaremos las columnas numéricas
columnas_numericas = ['Longitud', 'Diámetro', 'Altura', 'Peso_entero', 'Peso_manzana', 'Peso_vísceras', 'Peso_concha', 'Anillos']
df[columnas_numericas] = scaler.fit_transform(df[columnas_numericas])
```

```python
# Calculamos el rango intercuartílico (IQR) para cada columna numérica
Q1 = df[columnas_numericas].quantile(0.25)
Q3 = df[columnas_numericas].quantile(0.75)
IQR = Q3 - Q1

# Filtramos filas que están fuera del rango IQR
df_sin_outliers = df[~((df[columnas_numericas] < (Q1 - 1.5 * IQR)) | (df[columnas_numericas] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Imprimimos las filas que se eliminaron como outliers
filas_outliers = df[((df[columnas_numericas] < (Q1 - 1.5 * IQR)) | (df[columnas_numericas] > (Q3 + 1.5 * IQR))).any(axis=1)]
print("Filas eliminadas como outliers:")
print(filas_outliers)
```

```python
from sklearn.model_selection import train_test_split

# División de los datos para la tarea de regresión
X_regresion = df.drop(['Anillos'], axis=1)  # Características (todas excepto 'Anillos')
y_regresion = df['Anillos']  # Variable objetivo ('Anillos')

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_regresion, y_regresion, test_size=0.2, random_state=42)

# División de los datos para la tarea de clasificación
X_clasificacion = df.drop(['Sexo_F', 'Sexo_I', 'Sexo_M'], axis=1)  # Características (todas excepto las columnas de sexo codificadas)
y_clasificacion = df[['Sexo_F', 'Sexo_I', 'Sexo_M']]  # Variables objetivo (columnas de sexo codificadas)

X_train_clas, X_test_clas, y_train_clas, y_test_clas = train_test_split(X_clasificacion, y_clasificacion, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Inicializaremos y entrenaremos el modelo KNN para regresión
knn_regresion = KNeighborsRegressor()
knn_regresion.fit(X_train_reg, y_train_reg)

# Predecimos en el conjunto de prueba
y_pred_reg = knn_regresion.predict(X_test_reg)

# Calculamos el error cuadrático medio (MSE)
mse_reg = mean_squared_error(y_test_reg, y_pred_reg)
print("Error cuadrático medio (MSE) para regresión:", mse_reg)
```

El desempeño del modelo KNN para regresión en la tarea de predecir la edad del abulón fue evaluado utilizando el Error cuadrático medio (MSE). El MSE calculado fue de aproximadamente 5.13. Esto indica que, en promedio, las predicciones del modelo tienen un error cuadrático medio de 5.13 unidades en relación con los valores reales de la edad del abulón. En general, un MSE bajo sugiere que el modelo puede realizar buenas predicciones, aunque el contexto específico de la aplicación determinará si este nivel de precisión es adecuado para el propósito previsto.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Inicializaremos y entrenaremos el modelo KNN para clasificación
knn_clasificacion = KNeighborsClassifier()
knn_clasificacion.fit(X_train_clas, y_train_clas)

# Predeciremos en el conjunto de prueba
y_pred_clas = knn_clasificacion.predict(X_test_clas)

# Calcularemos la precisión del modelo
accuracy_clas = accuracy_score(y_test_clas, y_pred_clas)
print("Precisión del modelo KNN para clasificación:", accuracy_clas)

# Mostraremos el reporte de clasificación
print("\nReporte de clasificación:")
print(classification_report(y_test_clas, y_pred_clas))
```

El desempeño del modelo KNN para clasificación se evaluó utilizando la precisión del modelo y el reporte de clasificación. La precisión del modelo fue aproximadamente del 49.76%. Esto indica que el modelo clasificó correctamente alrededor del 49.76% de las muestras en el conjunto de prueba.

Al analizar el reporte de clasificación, observamos que las métricas de precisión, recall y f1-score varían para cada clase (0, 1 y 2), lo que indica cómo el modelo se desempeñó en la clasificación de cada clase en comparación con las otras clases. Además, la métrica de precisión promedio ponderada (weighted avg) y la métrica de precisión promedio macro (macro avg) proporcionan una visión general del desempeño general del modelo en todas las clases. En este caso, tanto la precisión promedio ponderada como la precisión promedio macro son aproximadamente del 57%, lo que sugiere que el modelo tiene un desempeño moderado en la clasificación de las clases.

La métrica samples avg proporciona la precisión promedio sobre las muestras reales, que es del 50%, lo que indica que el modelo tiene un rendimiento medio en la clasificación general de las muestras.

¿Podrías haber utilizado la regresión lineal y la regresión KNN para resolver el problema de regresión?

Sí, tanto la regresión lineal como la regresión KNN son técnicas que pueden aplicarse para resolver problemas de regresión. Sin embargo, es importante considerar las características del conjunto de datos y las suposiciones de cada modelo al elegir la técnica más adecuada.

Regresión lineal: Es una técnica que asume una relación lineal entre las características y la variable objetivo. Se puede utilizar cuando hay una relación lineal entre las características y la variable objetivo, y cuando se cumplen las suposiciones de la regresión lineal, como la normalidad de los residuos y la independencia de estos.
Regresión KNN (K-Nearest Neighbors): Es un algoritmo de aprendizaje supervisado no paramétrico que se utiliza tanto para problemas de regresión como de clasificación. Funciona encontrando los "k" puntos de datos más cercanos en el espacio de características y prediciendo el valor de la variable objetivo basado en la media (en el caso de regresión) de estos puntos. Puede ser útil cuando no se conoce la forma de la relación entre las características y la variable objetivo, y cuando no se cumplen las suposiciones de la regresión lineal.
En resumen, tanto la regresión lineal como la regresión KNN podrían haber sido opciones viables para resolver el problema de regresión, y la elección entre ellas dependería de la naturaleza del conjunto de datos y de si se cumplen las suposiciones de la regresión lineal.



¿Podrías haber utilizado la regresión lineal para el problema de clasificación?


No, la regresión lineal no es una técnica adecuada para resolver problemas de clasificación. La regresión lineal se utiliza para predecir valores continuos, mientras que los problemas de clasificación implican predecir clases o categorías discretas. La regresión lineal no está diseñada para manejar la naturaleza discreta de las variables objetivo en problemas de clasificación.

Para los problemas de clasificación, se suelen utilizar técnicas específicas como la regresión logística, los árboles de decisión, las máquinas de vectores de soporte (SVM) o algoritmos de vecinos más cercanos (KNN) para clasificación, entre otros. Estos algoritmos están diseñados para manejar la naturaleza discreta de las variables objetivo y son más apropiados para problemas de clasificación.
